# I. Streams

There are 3 major streams that we have discussed:
- iostream
- sstream
- fstream

## Input and Output Functionality
Each comes with a pair of input and output form
- iostream
    - cin : input stream for console
    - cout: output stream for console
- sstream
    - istringstream: input stream for strings
    - ostringstream: output stream for strings
- fstream
    - ifstream: input stream for files
    - ofstream: output stream for files
    
## Insertion and Extraction Operators
The insertion and extraction operators facilitate moving space separated data to and from the streams.
